```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the data for twitter training 
df = pd.read_csv('twitter_training.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# Set correct column names
df.columns = ['tweet_id', 'topic', 'sentiment', 'tweet']
```

```python
# Display the first few rows with correct column names
df.head()
```

```python
# Display unique values in the sentiment column to understand the labels
df['sentiment'].unique()
```

```python
# Drop rows with missing values
df.dropna(inplace=True)

# Display the cleaned data
df.head()
```

```python
# Sentiment Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='sentiment', data=df, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Topic-based Sentiment Distribution
plt.figure(figsize=(18, 8))
sns.countplot(x='topic', hue='sentiment', data=df, palette='coolwarm')
plt.title('Topic-based Sentiment Distribution')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
